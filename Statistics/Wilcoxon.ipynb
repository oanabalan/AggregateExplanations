```python
import openpyxl
import pandas as pd
import numpy as np
from scipy.stats import wilcoxon
import pprint
import pandas as pd

dataframe = openpyxl.load_workbook("ROC_Water.xlsx")
sheet = dataframe.active
names = ['R_AV', 'R_FI', 'R_AVFI', 'R_A', 'R_AFI', 'S_A', 'S_AFI', 'SR_A', 'SR_AFI', 'KNN']
list = []
for i in range(10):
    list_classifier = []
    list.append(list_classifier)

for row in range(1, sheet.max_row):
    k = 0
    for col in sheet.iter_cols(1, sheet.max_column):
        list[k].append(col[row].value)
        k = k + 1


matrix = [[0 for col in range(9)] for row in range(9)]
knn = [0] * 9

for current in range(9):
    p = []
    first = []
    second = []
    for i in range(current, current + 1):
        for j in range(0, 9):
            if i != j: 
                print(i, names[i], j, names[j])
                res = wilcoxon(list[i], list[j])
                print(res.pvalue)
                p.append(res.pvalue)
                first.append(i)
                second.append(j)
                if res.pvalue < 0.05:
                    print("Different")
                print("----------")
        
    
    for i in range(len(p) - 1):
        for j in range(i + 1, len(p)):
            if p[i] > p[j]:
                aux = p[i]
                p[i] = p[j]
                p[j] = aux
                aux = first[i]
                first[i] = first[j]
                first[j] = aux
                aux = second[i]
                second[i] = second[j]
                second[j] = aux
    
    print(p)
    print(first)
    print(second)
    
    alpha = []
    k = 0
    for i in range(8, 0, -1):
        a = 0.05 / i
        alpha.append(a)
        k = k + 1
    print(alpha)
    
    for i in range(8):
        if p[i] < alpha[i]:
            print(first[i], second[i], names[first[i]], names[second[i]], " Different")
            matrix[first[i]][second[i]] = p[i]
        else:
            break
    print("=======================")    

pprint.pprint(matrix)


p = []
first = []
second = []
i = 9
for j in range(0, 9):
    print(i, names[i], j, names[j])
    res = wilcoxon(list[i], list[j])
    print(res.pvalue)
    p.append(res.pvalue)
    first.append(i)
    second.append(j)
    if res.pvalue < 0.05:
        print("Different")
    print("----------")
        
    
for i in range(len(p) - 1):
    for j in range(i + 1, len(p)):
        if p[i] > p[j]:
            aux = p[i]
            p[i] = p[j]
            p[j] = aux
            aux = first[i]
            first[i] = first[j]
            first[j] = aux
            aux = second[i]
            second[i] = second[j]
            second[j] = aux
    
print(p)
print(first)
print(second)
    
alpha = []
k = 0
for i in range(8, 0, -1):
    a = 0.05 / i
    alpha.append(a)
    k = k + 1
print(alpha)
    
for i in range(8):
    if p[i] < alpha[i]:
        print(first[i], second[i], names[first[i]], names[second[i]], " Different")
        knn[second[i]] = p[i]
    else:
        break
   
print(knn)

matrix = np.append(matrix, [knn], axis=0)
df = pd.DataFrame(matrix)
df.to_excel(excel_writer = "Wilcoxon_ROC_Water.xlsx")

```
