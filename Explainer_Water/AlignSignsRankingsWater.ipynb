```python
import openpyxl
import pandas as pd
import numpy as np
from numpy.linalg import norm

feature_names = ['ph',	'Hardness',	'Solids',	'Chloramines',	'Sulfate',	'Conductivity',	'Organic_carbon',	'Trihalomethanes',	'Turbidity']
feature_names2 = ['ph',	'Hardness',	'Solids',	'Chloramines',	'Sulfate',	'Conductivity',	'Organic_carbon',	'Trihalomethanes',	'Turbidity',
                 'ph',	'Hardness',	'Solids',	'Chloramines',	'Sulfate',	'Conductivity',	'Organic_carbon',	'Trihalomethanes',	'Turbidity']
noOfIterations = 20

def convertSign(value):
    if value == "positive":
        return 1
    elif value == "negative":
        return 0
    else:
        return -1

def generateSignsRankings(feature_names):
    dataframe = openpyxl.load_workbook("outputWaterKernelShapLOOCV_1.xlsx")
    sheet = dataframe.active
   
    list = []
    for i in range(len(feature_names)):
        list_feature = []
        list.append(list_feature)
    
    df_KernelShapRankings = pd.DataFrame()
    
    for i, row in enumerate(sheet):
        step = 7
        if i == 0: 
            continue
        for k in range(len(feature_names)):
            if k == 0:
                list[k].append(row[4].value)
            else:
                list[k].append(row[step].value)
                step = step + 3
    dict = {}
    for i in range(len(feature_names)):
        new_data = {feature_names[i]: list[i]} 
        dict.update(new_data)
        
    df_KernelShapRankings = pd.DataFrame(dict)
    
    for i in range(len(df_KernelShapRankings)):
        list = []
        for j in range(len(feature_names)):
            count = 0
            for k in range(len(feature_names)):
               if df_KernelShapRankings.iloc[i, k] > df_KernelShapRankings.iloc[i, j]:
                    count = count + 1
            count = count + 1
            list.append(count)
        df_KernelShapRankings.loc[i] = list

    list = []
    for i in range(len(feature_names)):
        list_feature = []
        list.append(list_feature)
    
    df_KernelShapSigns = pd.DataFrame()
    
    for i, row in enumerate(sheet):
        step = 6
        if i == 0: 
            continue
        for k in range(len(feature_names)):
            if k == 0:
                list[k].append(convertSign(row[3].value))
            else:
                list[k].append(convertSign(row[step].value))
                step = step + 3
    dict = {}
    for i in range(len(feature_names)):
        new_data = {feature_names[i]: list[i]} 
        dict.update(new_data)
        
    df_KernelShapSigns = pd.DataFrame(dict)
  
    df_KernelShap = pd.concat([df_KernelShapRankings, df_KernelShapSigns], axis = 1)
    
    df_KernelShap.to_excel("signs_rankingsWaterKernelShap.xlsx")
   
    #################################################################
    dataframe = openpyxl.load_workbook("outputWaterTreeShapLOOCV_1.xlsx")
    sheet = dataframe.active
   
    list = []
    for i in range(len(feature_names)):
        list_feature = []
        list.append(list_feature)
    
    df_TreeShapRankings = pd.DataFrame()
    
    for i, row in enumerate(sheet):
        step = 7
        if i == 0: 
            continue
        for k in range(len(feature_names)):
            if k == 0:
                list[k].append(row[4].value)
            else:
                list[k].append(row[step].value)
                step = step + 3          
          
    dict = {}
    for i in range(len(feature_names)):
        new_data = {feature_names[i]: list[i]} 
        dict.update(new_data)
        
    df_TreeShapRankings = pd.DataFrame(dict)
   
    for i in range(len(df_TreeShapRankings)):
        list = []
        for j in range(len(feature_names)):
            count = 0
            for k in range(len(feature_names)):
               if df_TreeShapRankings.iloc[i, k] > df_TreeShapRankings.iloc[i, j]:
                    count = count + 1
            count = count + 1
            list.append(count)
        df_TreeShapRankings.loc[i] = list


    list = []
    for i in range(len(feature_names)):
        list_feature = []
        list.append(list_feature)
    
    df_TreeShapSigns = pd.DataFrame()
    
    for i, row in enumerate(sheet):
        step = 6
        if i == 0: 
            continue
        for k in range(len(feature_names)):
            if k == 0:
                list[k].append(convertSign(row[3].value))
            else:
                list[k].append(convertSign(row[step].value))
                step = step + 3          
          
    dict = {}
    for i in range(len(feature_names)):
        new_data = {feature_names[i]: list[i]} 
        dict.update(new_data)
        
    df_TreeShapSigns = pd.DataFrame(dict)

    df_TreeShap = pd.concat([df_TreeShapRankings, df_TreeShapSigns], axis = 1)

    df_TreeShap.to_excel("signs_rankingsWaterTreeShap.xlsx")    

    # ############################################################################
    dataframe = openpyxl.load_workbook("outputWaterLIMELOOCV_20.xlsx")
    sheet = dataframe.active
    
    list = []
    for i in range(len(feature_names)):
        list_feature = []
        list.append(list_feature)
    
    df_LIMERankings = pd.DataFrame()
    
    for i, row in enumerate(sheet):
        step = 11
        if i == 0: 
            continue
        for k in range(len(feature_names)):
            if k == 0:
                list[k].append(row[6].value)
            else:
                list[k].append(row[step].value)
                step = step + 5            
            
    dict = {}
    for i in range(len(feature_names)):
        new_data = {feature_names[i]: list[i]} 
        dict.update(new_data)
        
    df_LIMERankings = pd.DataFrame(dict)
   
    for i in range(len(df_LIMERankings)):
        list = []
        for j in range(len(feature_names)):
            count = 0
            for k in range(len(feature_names)):
               if df_LIMERankings.iloc[i, k] > df_LIMERankings.iloc[i, j]:
                    count = count + 1
            count = count + 1
            list.append(count)
        df_LIMERankings.loc[i] = list


    list = []
    for i in range(len(feature_names)):
        list_feature = []
        list.append(list_feature)
    
    df_LIMESigns = pd.DataFrame()
    
    for i, row in enumerate(sheet):
        step = 10
        if i == 0: 
            continue
        for k in range(len(feature_names)):
            if k == 0:
                list[k].append(convertSign(row[5].value))
            else:
                list[k].append(convertSign(row[step].value))
                step = step + 5            
            
    dict = {}
    for i in range(len(feature_names)):
        new_data = {feature_names[i]: list[i]} 
        dict.update(new_data)
        
    df_LIMESigns = pd.DataFrame(dict)
    df_LIME = pd.concat([df_LIMERankings, df_LIMESigns], axis = 1)

    df_LIME.to_excel("signs_rankingsWaterLIME.xlsx")
    ################################################################
    

def stackRowsColumns(M, noOfRows, noOfColumns):
    pos = [0] * noOfRows
    for i in range (0, noOfRows):
        pos[i] = i
    
    for i in range(1, noOfRows):
        wsim = [0] * noOfRows
        for j in range(i, noOfRows):
            for k in range(0, i):
                A = M.loc[k]
                B = M.loc[j]
                wsim[j] = wsim[j] + (1 / ( i - k)) * np.dot(A,B)/(norm(A)*norm(B))
        max_value = max(wsim)
        max_index = wsim.index(max_value)
        aux = M.loc[max_index]
        M.loc[max_index] = M.loc[i]
        M.loc[i] = aux
        aux = pos[max_index]
        pos[max_index] = pos[i]
        pos[i] = aux

    
    # for i in range(1, noOfColumns):
    #     wsim = [0] * noOfColumns
    #     for j in range(i, noOfColumns):
    #         for k in range(0, i):
    #             A = M.loc[k]
    #             B = M.loc[j]
    #             wsim[j] = wsim[j] + (1 / ( i - k)) * np.dot(A,B)/(norm(A)*norm(B))
    #     max_value = max(wsim)
    #     max_index = wsim.index(max_value)
    #     aux = M.loc[max_index]
    #     M.loc[max_index] = M.loc[i]
    #     M.loc[i] = aux
    return pos, M

def cosineSimilarity(A, B):
    return np.dot(A,B)/(norm(A)*norm(B))

def similarityCase(M, i):
    s = 0
    k = i - 1
    for j in range (1, k + 1):
        s = s + cosineSimilarity(M.loc[i], M.loc[i - j]) * (1 / j)
    return s

def similarityMatrix(M, noOfRows):
    s = 0
    for i in range(1, noOfRows):
        s = s + similarityCase(M, i)
    s = s / (noOfRows - 1)
    return s

def GlobalAlign(MSP, MS, noOfRows):
    # if similarityMatrix(MSP, noOfRows) >  similarityMatrix(MS, noOfRows):
    #     max = similarityMatrix(MSP, noOfRows)
    #     min = similarityMatrix(MS, noOfRows)
    # else:
    #     max = similarityMatrix(MS, noOfRows)
    #     min = similarityMatrix(MSP, noOfRows)
    # return min / max
    return similarityMatrix(MSP, noOfRows) / similarityMatrix(MS, noOfRows)


def createMSP(MP, MS, noOfRows):
    MS_copy = MS.copy(deep = True)
    MP_copy = MP.copy(deep = True)
    posS, MS_copy = stackRowsColumns(MS_copy, noOfRows, 2 * len(feature_names))
    posP, MP_copy = stackRowsColumns(MP_copy, noOfRows, 2 * len(feature_names))
    MSP = pd.DataFrame(columns = feature_names2)
    for i in range(0, noOfRows):
        MSP.loc[posP[i]] = MS_copy.loc[posS[i]]
    return GlobalAlign(MSP, MS_copy, noOfRows)

generateSignsRankings(feature_names)
df_KernelShap = pd.read_excel('signs_rankingsWaterKernelShap.xlsx')
df_KernelShap = df_KernelShap[df_KernelShap.columns[1:]]
df_TreeShap = pd.read_excel('signs_rankingsWaterTreeShap.xlsx')
df_TreeShap = df_TreeShap[df_TreeShap.columns[1:]]
df_LIME = pd.read_excel('signs_rankingsWaterLIME.xlsx')
df_LIME = df_LIME[df_LIME.columns[1:]]


results = pd.DataFrame(columns = ['LL', 'LS', 'SL', 'SS'])
noOfInstances = df_KernelShap.shape[0]
for instance in range(0, noOfInstances):
    print(instance)
    result = []
    data_Shap = []
    for i in range(int(noOfIterations / 2)):
        data_Shap.append(df_KernelShap.loc[instance].values.flatten().tolist())
        data_Shap.append(df_TreeShap.loc[instance].values.flatten().tolist())
    M_Shap  = pd.DataFrame(data_Shap, columns = feature_names2)
    
    data_LIME = []
    for i in range(noOfIterations):
        data_LIME.append(df_LIME.loc[instance * noOfIterations + i].values.flatten().tolist())
    M_LIME  = pd.DataFrame(data_LIME, columns = feature_names2)
    
    MP = M_LIME
    MS = M_LIME
    result.append(createMSP(MP, MS, noOfIterations))
    MP = M_LIME
    MS = M_Shap
    result.append(createMSP(MP, MS, noOfIterations))
    MP = M_Shap
    MS = M_LIME
    result.append(createMSP(MP, MS, noOfIterations))
    MP = M_Shap
    MS = M_Shap
    result.append(createMSP(MP, MS, noOfIterations))
    print(result)
    results.loc[instance] = result


results.to_excel("resultsSignsRankingsWaterMSPMS_NoMinMax.xlsx")
```
