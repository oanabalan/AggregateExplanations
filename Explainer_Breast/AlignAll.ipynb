```python
import openpyxl
import pandas as pd
import numpy as np
from numpy.linalg import norm

feature_names = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI','DiabetesPedigreeFunction', 'Age']
noOfIterations = 1

def generateRankings(feature_names):
    dataframe = openpyxl.load_workbook("outputDiabetesKernelShapLOOCV_1.xlsx")
    sheet = dataframe.active
   
    list = []
    for i in range(len(feature_names)):
        list_feature = []
        list.append(list_feature)
    
    df_KernelShap = pd.DataFrame()
    
    for i, row in enumerate(sheet):
        step = 7
        if i == 0: 
            continue
        for k in range(len(feature_names)):
            if k == 0:
                list[k].append(row[4].value)
            else:
                list[k].append(row[step].value)
                step = step + 3
    dict = {}
    for i in range(len(feature_names)):
        new_data = {feature_names[i]: list[i]} 
        dict.update(new_data)
        
    df_KernelShap = pd.DataFrame(dict)
    
    for i in range(len(df_KernelShap)):
        list = []
        for j in range(len(feature_names)):
            count = 0
            for k in range(len(feature_names)):
               if df_KernelShap.iloc[i, k] > df_KernelShap.iloc[i, j]:
                    count = count + 1
            count = count + 1
            list.append(count)
        df_KernelShap.loc[i] = list
    
    df_KernelShap.to_excel("rankingsKernelShapAll.xlsx")
   
    #################################################################
    dataframe = openpyxl.load_workbook("outputDiabetesTreeShapLOOCV_1.xlsx")
    sheet = dataframe.active
   
    list = []
    for i in range(len(feature_names)):
        list_feature = []
        list.append(list_feature)
    
    df_TreeShap = pd.DataFrame()
    
    for i, row in enumerate(sheet):
        step = 7
        if i == 0: 
            continue
        for k in range(len(feature_names)):
            if k == 0:
                list[k].append(row[4].value)
            else:
                list[k].append(row[step].value)
                step = step + 3          
          
    dict = {}
    for i in range(len(feature_names)):
        new_data = {feature_names[i]: list[i]} 
        dict.update(new_data)
        
    df_TreeShap = pd.DataFrame(dict)
   
    for i in range(len(df_TreeShap)):
        list = []
        for j in range(len(feature_names)):
            count = 0
            for k in range(len(feature_names)):
               if df_TreeShap.iloc[i, k] > df_TreeShap.iloc[i, j]:
                    count = count + 1
            count = count + 1
            list.append(count)
        df_TreeShap.loc[i] = list
    
    df_TreeShap.to_excel("rankingsTreeShapAll.xlsx")    

    ############################################################################
    dataframe = openpyxl.load_workbook("outputDiabetesLIMELOOCV_1.xlsx")
    sheet = dataframe.active
   
    list = []
    for i in range(len(feature_names)):
        list_feature = []
        list.append(list_feature)
    
    df_LIME = pd.DataFrame()
    
    for i, row in enumerate(sheet):
        step = 11
        if i == 0: 
            continue
        for k in range(len(feature_names)):
            if k == 0:
                list[k].append(row[6].value)
            else:
                list[k].append(row[step].value)
                step = step + 5            
            
    dict = {}
    for i in range(len(feature_names)):
        new_data = {feature_names[i]: list[i]} 
        dict.update(new_data)
        
    df_LIME = pd.DataFrame(dict)
   
    for i in range(len(df_LIME)):
        list = []
        for j in range(len(feature_names)):
            count = 0
            for k in range(len(feature_names)):
               if df_LIME.iloc[i, k] > df_LIME.iloc[i, j]:
                    count = count + 1
            count = count + 1
            list.append(count)
        df_LIME.loc[i] = list
    
    df_LIME.to_excel("rankingsLIMEAll.xlsx")
    ################################################################
    dataframe = openpyxl.load_workbook("outputDiabetesAnchorLOOCV_1.xlsx")
    sheet = dataframe.active
    
    list = []
    for i in range(len(feature_names)):
        list_feature = []
        list.append(list_feature)
    
    df_Anchor = pd.DataFrame()
    
    for i, row in enumerate(sheet):
        step = 7
        if i == 0: 
            continue
        for k in range(len(feature_names)):
            if k == 0:
                val = row[4].value
                if val == -1:
                    val = 0
                list[k].append(val)
            else:
                val = row[step].value
                if val == -1:
                   val = 0 
                list[k].append(val)
                step = step + 3
        
    dict = {}
    for i in range(len(feature_names)):
        new_data = {feature_names[i]: list[i]} 
        dict.update(new_data)
        
    df_Anchor = pd.DataFrame(dict)
  
    for i in range(len(df_Anchor)):
        list = []
        for j in range(len(feature_names)):
            count = 0
            for k in range(len(feature_names)):
               if df_Anchor.iloc[i, k] > df_Anchor.iloc[i, j]:
                    count = count + 1
            count = count + 1
            list.append(count)
        df_Anchor.loc[i] = list
    
    df_Anchor.to_excel("rankingsAnchorAll.xlsx")

def stackRowsColumns(M, noOfRows, noOfColumns):
    print("row")
    pos = [0] * noOfRows
    for i in range (0, noOfRows):
        pos[i] = i
    
    for i in range(1, noOfRows):
        wsim = [0] * noOfRows
        for j in range(i, noOfRows):
            for k in range(0, i):
                A = M.loc[k]
                B = M.loc[j]
                wsim[j] = wsim[j] + (1 / ( i - k)) * np.dot(A,B)/(norm(A)*norm(B))
        max_value = max(wsim)
        max_index = wsim.index(max_value)
        aux = M.loc[max_index]
        M.loc[max_index] = M.loc[i]
        M.loc[i] = aux
        aux = pos[max_index]
        pos[max_index] = pos[i]
        pos[i] = aux
    print("col")
    wsim = [0] * noOfColumns
    for i in range(1, noOfColumns):
        for j in range(i, noOfColumns):
            wsim[j] = 0
            for k in range(0, i):
                A = M.loc[k]
                B = M.loc[j]
                wsim[j] = wsim[j] + (1 / ( i - k)) * np.dot(A,B)/(norm(A)*norm(B))
        max_value = max(wsim)
        max_index = wsim.index(max_value)
        aux = M.loc[max_index]
        M.loc[max_index] = M.loc[i]
        M.loc[i] = aux
    return pos, M

def cosineSimilarity(A, B):
    return np.dot(A,B)/(norm(A)*norm(B))

def similarityCase(M, i):
    s = 0
    k = i - 1
    for j in range (1, k + 1):
        s = s + cosineSimilarity(M.loc[i], M.loc[i - j]) * (1 / j)
    return s

def similarityMatrix(M, noOfRows):
    s = 0
    for i in range(1, noOfRows):
        s = s + similarityCase(M, i)
    s = s / (noOfRows - 1)
    return s

def GlobalAlign(MSP, MS, noOfRows):
    return similarityMatrix(MSP, noOfRows) / similarityMatrix(MS, noOfRows)


def createMSP(MP, MS, noOfRows):
    MS_copy = MS.copy(deep = True)
    MP_copy = MP.copy(deep = True)
    posS, MS_copy = stackRowsColumns(MS_copy, noOfRows, len(feature_names))
    print(posS)
    posP, MP_copy = stackRowsColumns(MP_copy, noOfRows, len(feature_names))
    print(posP)
    MSP = pd.DataFrame(columns = feature_names)
    for i in range(0, noOfRows):
        MSP.loc[posP[i]] = MS_copy.loc[posS[i]]
    return GlobalAlign(MSP, MS_copy, noOfRows)

#generateRankings(feature_names)
df_KernelShap = pd.read_excel('rankingsKernelShapAll.xlsx')
df_KernelShap = df_KernelShap[df_KernelShap.columns[1:]]
df_TreeShap = pd.read_excel('rankingsTreeShapAll.xlsx')
df_TreeShap = df_TreeShap[df_TreeShap.columns[1:]]
df_LIME = pd.read_excel('rankingsLIMEAll.xlsx')
df_LIME = df_LIME[df_LIME.columns[1:]]
df_Anchor = pd.read_excel('rankingsAnchorAll.xlsx')
df_Anchor = df_Anchor[df_Anchor.columns[1:]]


results = pd.DataFrame(columns = ['AA', 'AL', 'AKS', 'ATS', 'LA', 'LL', 'LKS', 'LTS', 'KSA', 'KSL', 'KSKS', 'KSTS', 'TSA', 'TSL', 'TSKS', 'TSTS'])
noOfInstances = df_KernelShap.shape[0]

result = []
data_TreeShap = []
for instance in range(noOfInstances):
    data_TreeShap.append(df_TreeShap.loc[instance].values.flatten().tolist())
M_TreeShap  = pd.DataFrame(data_TreeShap, columns = feature_names)

data_KernelShap = []
for instance in range(noOfInstances):
    data_KernelShap.append(df_KernelShap.loc[instance].values.flatten().tolist())
M_KernelShap  = pd.DataFrame(data_TreeShap, columns = feature_names)

data_LIME = []
for instance in range(noOfInstances):
    data_LIME.append(df_LIME.loc[instance].values.flatten().tolist())
M_LIME  = pd.DataFrame(data_LIME, columns = feature_names)

data_Anchor = []
for instance in range(noOfInstances):
    data_Anchor.append(df_Anchor.loc[instance].values.flatten().tolist())
M_Anchor  = pd.DataFrame(data_Anchor, columns = feature_names)  

MP = M_Anchor
MS = M_Anchor
r = createMSP(MP, MS, noOfInstances)
print(r)
result.append(r)
MP = M_Anchor
MS = M_LIME
r = createMSP(MP, MS, noOfInstances)
print(r)
result.append(r)
MP = M_Anchor
MS = M_KernelShap
r = createMSP(MP, MS, noOfInstances)
print(r)
result.append(r)
MP = M_Anchor
MS = M_TreeShap
r = createMSP(MP, MS, noOfInstances)
print(r)
result.append(r)
MP = M_LIME
MS = M_Anchor
r = createMSP(MP, MS, noOfInstances)
print(r)
result.append(r)
MP = M_LIME
MS = M_LIME
r = createMSP(MP, MS, noOfInstances)
print(r)
result.append(r)
MP = M_LIME
MS = M_KernelShap
r = createMSP(MP, MS, noOfInstances)
print(r)
result.append(r)
MP = M_LIME
MS = M_TreeShap
r = createMSP(MP, MS, noOfInstances)
print(r)
result.append(r)
MP = M_KernelShap
MS = M_Anchor
r = createMSP(MP, MS, noOfInstances)
print(r)
result.append(r)
MP = M_KernelShap
MS = M_LIME
r = createMSP(MP, MS, noOfInstances)
print(r)
result.append(r)
MP = M_KernelShap
MS = M_KernelShap
r = createMSP(MP, MS, noOfInstances)
print(r)
result.append(r)
MP = M_KernelShap
MS = M_TreeShap
r = createMSP(MP, MS, noOfInstances)
print(r)
result.append(r)
MP = M_TreeShap
MS = M_Anchor
r = createMSP(MP, MS, noOfInstances)
print(r)
result.append(r)
MP = M_TreeShap
MS = M_LIME
r = createMSP(MP, MS, noOfInstances)
print(r)
result.append(r)
MP = M_TreeShap
MS = M_KernelShap
r = createMSP(MP, MS, noOfInstances)
print(r)
result.append(r)
MP = M_TreeShap
MS = M_TreeShap
r = createMSP(MP, MS, noOfInstances)
print(r)
result.append(r)
print(result)
results.loc[0] = result

print(results)
results.to_excel("resultsAll.xlsx")
```
